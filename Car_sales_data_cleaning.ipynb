Price is cleaned

```python
import pandas as pd
import re

# Function to clean and extract the values
def clean_and_extract(row):
    original = str(row['Price'])

    # Extract all dollar amounts
    prices = re.findall(r'\$\d{1,3}(?:,\d{3})*(?:\.\d{2})?', original)
    # Convert prices to integers
    prices = [int(price.replace('$', '').replace(',', '')) for price in prices]

    msrp, price, price_drop = None, None, None

    if 'MSRP' in original and len(prices) == 2:
        msrp, price = prices
        price_drop = price - msrp
    elif 'price drop' in original and len(prices) == 2:
        price, price_drop = prices
        msrp = price - price_drop
    elif len(prices) == 1:
        price = prices[0]

    return pd.Series([msrp, price, price_drop], index=['MSRP', 'Selling Price', 'Price Drop'])

# Apply the function to the DataFrame
df[['MSRP', 'Selling Price', 'Price Drop']] = df.apply(clean_and_extract, axis=1)

```

Vehicle Review info is cleaned

```python
import ast
# Function to parse string to dictionary
def parse_review_info(review_info):
    try:
        return ast.literal_eval(review_info)
    except (ValueError, SyntaxError) as e:
        print(f"Error parsing review info: {e}")
        return {}

# Parse the 'Vehicle Reviews Info' column to extract key-value pairs
df['Parsed Reviews'] = df['Vehicle Reviews Info'].apply(parse_review_info)

# Create separate columns for each key in the dictionary
for key in ['Comfort', 'Interior', 'Performance', 'Value', 'Exterior', 'Reliability']:
    df[key] = df['Parsed Reviews'].apply(lambda x: x.get(key, None))

# Drop the 'Parsed Reviews' column as it's no longer needed
df.drop(columns=['Parsed Reviews'], inplace=True)
```
